# 1. Login to your Instagram Handle
Submit with sample username and password.

```python
#login function
def login(un, pw):
    driver.get('https://www.instagram.com/')
    username = wait.until(EC.presence_of_element_located((By.NAME, 'username')))
    username.send_keys(un)
    password = wait.until(EC.presence_of_element_located((By.NAME, 'password')))
    password.send_keys(pw)
    login = driver.find_element_by_xpath('//button[contains(@type, "submit")]')
    login.submit()
```

```python
un = 'SAMPLE USERNAME'
pw = 'SAMPLE PASSWORD'

login(un, pw)
```

# 2. Type for “food” in search bar and print all the names of the Instagram Handles that are displayed in list after typing “food”
Note : Make sure to avoid printing hashtags.

```python
#search function
def search_keyword(keyword):
    search = driver.find_element_by_class_name('TqC_a')
    search.click()
    search = driver.find_element_by_xpath('//input[contains(@placeholder, "Search")]')
    search.send_keys(keyword)
    search.send_keys(Keys.RETURN)
    wait = WebDriverWait(driver, 10)
    handles = wait.until(EC.presence_of_all_elements_located((By.XPATH, '//div[contains(@class, "fuqBx")]//div[contains(@class, "uL8Hv")]')))
    for handle in handles:
        text = handle.get_attribute('innerText')
        if(text[0] != '#'):
            print(text)
    driver.find_element_by_class_name('coreSpriteSearchClear').click()
```

```python
keyword = 'food'

search_keyword(keyword)
```

# 4. Follow/Unfollow given handle - 
1. Open the Instagram Handle of “So Delhi”

3. After following, unfollow the instagram handle. Print a message if you have already unfollowed.

```python
#function to follow a profile
def follow_profile(profile):
    open_profile(profile)
    driver.find_element_by_class_name('_6VtSN').click()
    print('Followed So Delhi!')
    driver.back()
```

```python
#function to unfollow a profile
def unfollow_profile(profile):
    open_profile(profile)
    driver.find_element(By.XPATH, '//button[contains(@class, "_6VtSN")]').click()
    driver.find_element_by_class_name('-Cab_').click()
    print('Unfollowed So Delhi!')
    driver.back()
```

# 5. Like/Unlike posts
1. Liking the top 30 posts of the ‘dilsefoodie'. Print message if you have already liked it.
2. Unliking the top 30 posts of the ‘dilsefoodie’. Print message if you have already unliked it.

```python
#function to like posts
def like_posts(profile):
    open_profile(profile)
    sleep(3)
    current_end = driver.execute_script('return document.body.scrollHeight;') 
    driver.execute_script('window.scrollTo(0, arguments[0]);', current_end)
    sleep(3)
    for i in range(30):
        driver.find_elements_by_class_name('_9AhH0')[i].click()
        sleep(3)
        driver.find_element_by_xpath('//span[contains(@class, "fr66n")]//button').click()
        driver.back()
    print('Liked 30 posts!')
```

```python
#function to unlike posts
def unlike_posts(profile):
    open_profile(profile)
    sleep(3)
    current_end = driver.execute_script('return document.body.scrollHeight;') 
    driver.execute_script('window.scrollTo(0, arguments[0]);', current_end)
    sleep(3)
    for i in range(30):
        driver.find_elements_by_class_name('_9AhH0')[i].click()
        sleep(3)
        driver.find_element_by_xpath('//span[contains(@class, "fr66n")]//button').click()
        driver.back()
    print('Unliked 30 posts!')
```

```python
profile = 'dilsefoodie'

like_posts(profile)
unlike_posts(profile)
```

# Following profiles as required by the question

```python
#function to follow profiles listed when searching with keyword
def follow_profiles_with_keyword(keyword):
    search = driver.find_element_by_class_name('TqC_a')
    search.click()
    search = driver.find_element_by_xpath('//input[contains(@placeholder, "Search")]')
    search.send_keys(keyword)
    search.send_keys(Keys.RETURN)
    wait = WebDriverWait(driver, 10)
    handles = wait.until(EC.presence_of_all_elements_located((By.XPATH, '//div[contains(@class, "fuqBx")]//div[contains(@class, "uL8Hv")]')))
    profiles = []
    for handle in handles:
        text = handle.get_attribute('innerText')
        if(text[0] != '#'):
            profiles.append(text)
    for profile in profiles:
        driver.get(f'https://www.instagram.com/{profile}/')
        wait = WebDriverWait(driver, 10)
        try:
            follow = wait.until(EC.presence_of_element_located((By.CLASS_NAME, '_6VtSN')))
            follow.click()
        except:
            continue
```

# 6. Extract list of followers
1. Extract the usernames of the first 500 followers of ‘foodtalkindia’ and ‘sodelhi’.
2. Now print all the followers of “foodtalkindia” that you are following but those who don’t follow you.

## 1.

```python
#function to print and extract followers of a profile
def extract_followers(profile):     
    open_profile(profile)
    wait = WebDriverWait(driver, 10)
    followers = wait.until(EC.presence_of_element_located((By.XPATH, '//a[contains(@class, "-nal3")]')))
    followers.click()
    wait = WebDriverWait(driver, 10)
    temp = wait.until(EC.presence_of_element_located((By.CLASS_NAME, 'isgrP')))
    for i in range(41):
        driver.execute_script('document.getElementsByClassName("isgrP")[0].scrollTop += 1000;')
        sleep(4)
    followers = set()
    usernames = driver.find_elements_by_class_name('FPmhX')
    for i in range(500):
        followers.add(usernames[i].get_attribute('innerHTML'))
        print(usernames[i].get_attribute('innerHTML'))
    driver.back()
    return followers
```

```python
profile1 = 'foodtalkindia'
profile2 = 'sodelhi'

driver.get('https://www.instagram.com/')
print('First 500 followers of foodtalkindia:')
profile1 = extract_followers(profile1)
print()
driver.get('https://www.instagram.com/')
print('First 500 followers of sodelhi:')
profile2 = extract_followers(profile2)
```

```python
#function to print followers as per given conditions
def print_followers(profile):
    my_followers = set()
    driver.find_elements_by_class_name('Y8-fY')[1].click()

    if((int)(driver.find_elements_by_class_name('g47SY')[1].get_attribute('innerHTML')) > 0):
        for i in range(5):
            driver.execute_script('document.getElementsByClassName("isgrP")[0].scrollTop += 1000;')
            sleep(4)

        followers = driver.find_elements_by_class_name('FPmhX') 
        for f in followers:
            my_followers.add(f.get_attribute('innerHTML'))
    wait = WebDriverWait(driver, 10)
    close = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, "wpO6b")))[2]
    close.click()
    
    my_following = set()
    driver.find_elements_by_class_name('Y8-fY')[2].click()

    if((int)(driver.find_elements_by_class_name('g47SY')[2].get_attribute('innerHTML')) > 0):
        for i in range(5):
            driver.execute_script('document.getElementsByClassName("isgrP")[0].scrollTop += 1000;')
            sleep(4)

    following = driver.find_elements_by_class_name('FPmhX') 
    for f in following:
        my_following.add(f.get_attribute('innerHTML'))
        
    res = ((profile & my_following) - my_followers)
    for user in res:
        print(user)
```

# 7. Check the story of ‘coding.ninjas’. Consider the following Scenarios and print error messages accordingly -
1. If You have already seen the story.
2. Or The user has no story.
3. Or View the story if not yet seen.

```python
#function to view the story of a profile
def view_story(profile):
    open_profile(profile)
    
    try:
        wait = WebDriverWait(driver, 10)
        story = wait.until(EC.presence_of_element_located((By.CLASS_NAME, "h5uC0")))
    except:
        print('User has no story!')
    else:
        sleep(2)
        canvas = driver.find_element_by_class_name('CfWVH')
        if(canvas.get_attribute('height') == '208'):
            print('Already seen the story!')
        else:
            story.click()
            print('Viewed the story!')
```

```python
profile = 'coding.ninjas'

view_story(profile)
```
